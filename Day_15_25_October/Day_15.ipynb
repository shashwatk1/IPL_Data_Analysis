## Functions

##### Head to Head and last 3 years

```python
def head2head(t1,t2,df = match):
    s1 = (df['team1'] == t1) & (df['team2'] == t2)
    s2 = (df['team1'] == t2) & (df['team2'] == t1)
    head = df.iloc[np.where(s1 | s2)]
    head = head.sort_values(by = 'season').reset_index().drop(['id','index'], axis=1)
    return head

y1 = deli.iloc[np.where(deli['year'] == 2017)[0]]
y2 = deli.iloc[np.where(deli['year'] == 2018)[0]]
y3 = deli.iloc[np.where(deli['year'] == 2019)[0]]
year3 = pd.DataFrame.append(y1,y2, ignore_index=True)
last3years = pd.DataFrame.append(year3,y3, ignore_index=True)
```

##### Head to Head Score Function

```python
def head(t1, t2):
  head_to_head = head2head(t1,t2)
  t1_win=0
  t2_win=0
  for i in head_to_head['winner']:
    if i == t1:
      t1_win+=1
    if i == t2:
      t2_win+=1
  t1_arr = []
  t2_arr = []
  year_arr = []
  for i in head_to_head[['season','winner']].iterrows():
      index, year, win = i[0], i[1][0], i[1][1]
      if year in year_arr:
          if win == t1:
              t1_arr[year_arr.index(year)]+=1  
          if win == t2:
              t2_arr[year_arr.index(year)]+=1  
      else:
          year_arr.append(year)
          if win == t1:
              t1_arr.append(1)
              t2_arr.append(0)
          if win == t2:
              t1_arr.append(0)
              t2_arr.append(1)
  win_by_year = pd.DataFrame([t1_arr, t2_arr], columns = year_arr, index = [t1, t2])
  print("Head to Head:",t1,'v/s',t2)
  print(t1+':', t1_win)
  print(t2+':', t2_win) 
  return win_by_year
```

##### Batsman Intimadation Score

```python
def bis_score(name, against, df):
    runs, norm_runs, bound_run, ball = [], [], [], []
    bdf = df[df['batsman']==name]
    if against != '':
        bdf = bdf[bdf['bowling_team']==against]
    for i in bdf['match_id'].unique():
        bdf2 = bdf[bdf['match_id']==i]
        runs.append(bdf2.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()['batsman_runs'][0])
        norm_runs.append(bdf2.groupby(['match_id'], sort = False)['norm_runs'].sum().reset_index()['norm_runs'][0])
        bound_run.append(bdf2.groupby(['match_id'], sort = False)['bound_runs'].sum().reset_index()['bound_runs'][0])
        ball.append(bdf2.groupby(['match_id'], sort = False)['ball'].count().reset_index()['ball'][0])
    ndf = pd.DataFrame({'tot_balls':ball,'bound_runs':bound_run,'norm_runs':norm_runs,'tot_runs':runs})  
    ndf['BIS'] = (ndf['tot_runs'] + ndf['bound_runs'] - (ndf['norm_runs'] * 0.5))/(ndf['tot_balls'])
    count, mean ,median, mini, maxi = ndf['BIS'].describe()[0], ndf['BIS'].describe()[1], ndf['BIS'].describe()[5], ndf['BIS'].describe()[3], ndf['BIS'].describe()[7]
    print('Matches Played:',count,'\tMean:',np.round_(mean, decimals = 3),'\tMedian:',np.round_(median, decimals = 3), '\tMinimum:',np.round_(mini, decimals = 3), '\tMaximum:',np.round_(maxi, decimals = 3))
```

##### Batsman boundary to total ratio

```python
def tot_bound_ratio(name, against, df):
    bdf = df[df['batsman']==name]
    if against!= '': 
        bdf = bdf[bdf['bowling_team']==against]
    bounds = []
    total = []
    for i in bdf['match_id'].unique():
        bdf2 = bdf[bdf['match_id']==i]
        total.append(bdf2.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()['batsman_runs'][0])
        bounds.append(bdf2.groupby(['match_id'], sort = False)['bound_runs'].sum().reset_index()['bound_runs'][0])
    df2 = pd.DataFrame({'tot_runs':total,'bound_runs':bounds})
    df2['ratio'] = df2['bound_runs']/df2['tot_runs']
    count, mean ,median, mini, maxi = df2['ratio'].describe()[0], df2['ratio'].describe()[1], df2['ratio'].describe()[5], df2['ratio'].describe()[3], df2['ratio'].describe()[7]
    print('Matches Played:',count,'\tMean:',np.round_(mean, decimals = 3),'\tMedian:',np.round_(median, decimals = 3), '\tMinimum:',np.round_(mini, decimals = 3), '\tMaximum:',np.round_(maxi, decimals = 3))
```

##### Balls to target runs

```python
def targ_runs(t1, t2, targ, df):
    df1 = df[df['batting_team'] == t1]
    df1 = df1[df1['bowling_team'] == t2].reset_index().drop(['index'], axis=1)
    runs = df1.groupby(['match_id'], sort = False)['total_runs'].sum().reset_index()
    runs = runs[runs['total_runs']>=targ]
    balls = []
    ball = 0
    for i in runs['match_id'].unique():
        dfi = df1[df1['match_id'] == i]
        df2 = dfi['total_runs'].values.tolist()
        dfex =  dfi['is_wide']
        runs = 0
        for j in df2:
            if runs <= targ:
                runs += j
                ball += 1 
        ball -= dfex.iloc[:ball].sum()
        balls.append(ball)
        ball = 0
    print(pd.DataFrame(balls).describe(),'\n')
    print(pd.DataFrame(balls).value_counts())
```

##### Bowler lethality score

```python
def bls_score(name, against, df):
    wick, dot, bound_count, bound_run, tot = [], [], [], [], []
    bdf = df[df['bowler']==name]
    if against != '':
        bdf = bdf[bdf['batting_team']==against]
    for i in bdf['match_id'].unique():
        bdf2 = bdf[bdf['match_id']==i]
        wick.append(bdf2.groupby(['match_id'], sort = False)['dismissal_by_bowler'].sum().reset_index()['dismissal_by_bowler'][0])
        dot.append(bdf2.groupby(['match_id'], sort = False)['is_dot'].sum().reset_index()['is_dot'][0])
        bound_count.append(bdf2.groupby(['match_id'], sort = False)['is_bound'].sum().reset_index()['is_bound'][0])
        bound_run.append(bdf2.groupby(['match_id'], sort = False)['bound_runs'].sum().reset_index()['bound_runs'][0])
        tot.append(bdf2.groupby(['match_id'], sort = False)['ball'].count().reset_index()['ball'][0])
    ndf = pd.DataFrame({'tot_balls':tot,'dot_balls':dot,'bound_count':bound_count,'bound_run':bound_run,'wickets':wick})  
    ndf['BLS'] = ndf['tot_balls'] - (ndf['dot_balls'] * 2)
    ndf = ndf[ndf['BLS']!=0]
    ndf['BLS'] = (ndf['wickets'] + (ndf['dot_balls'] * 2)- (ndf['bound_run']/ndf['bound_count'])) /(ndf['tot_balls'] - (ndf['dot_balls'] * 2))
    count, mean ,median, mini, maxi = ndf['BLS'].describe()[0], ndf['BLS'].describe()[1], ndf['BLS'].describe()[5], ndf['BLS'].describe()[3], ndf['BLS'].describe()[7]
    print('Matches Played:',count,'\tMean:',np.round_(mean, decimals = 3),'\tMedian:',np.round_(median, decimals = 3), '\tMinimum:',np.round_(mini, decimals = 3), '\tMaximum:',np.round_(maxi, decimals = 3))
```
