## Functions

##### Head to Head and last 3 years

```python
def head2head(t1,t2,df = match):
    s1 = (df['team1'] == t1) & (df['team2'] == t2)
    s2 = (df['team1'] == t2) & (df['team2'] == t1)
    head = df.iloc[np.where(s1 | s2)]
    head = head.sort_values(by = 'season').reset_index().drop(['id','index'], axis=1)
    return head

def last3years():
    y1 = deli.iloc[np.where(deli['year'] == 2017)[0]]
    y2 = deli.iloc[np.where(deli['year'] == 2018)[0]]
    y3 = deli.iloc[np.where(deli['year'] == 2019)[0]]
    year3 = pd.DataFrame.append(y1,y2, ignore_index=True)
    year3 = pd.DataFrame.append(year3,y3, ignore_index=True)
    return year3
```

##### Head to Head Score Function

```python
def head(t1, t2):
  head_to_head = head2head(t1,t2)
  t1_win=0
  t2_win=0
  for i in head_to_head['winner']:
    if i == t1:
      t1_win+=1
    if i == t2:
      t2_win+=1
  t1_arr = []
  t2_arr = []
  year_arr = []
  for i in head_to_head[['season','winner']].iterrows():
      index, year, win = i[0], i[1][0], i[1][1]
      if year in year_arr:
          if win == t1:
              t1_arr[year_arr.index(year)]+=1  
          if win == t2:
              t2_arr[year_arr.index(year)]+=1  
      else:
          year_arr.append(year)
          if win == t1:
              t1_arr.append(1)
              t2_arr.append(0)
          if win == t2:
              t1_arr.append(0)
              t2_arr.append(1)
  win_by_year = pd.DataFrame([t1_arr, t2_arr], columns = year_arr, index = [t1, t2])
  print("Head to Head:",t1,'v/s',t2)
  print(t1+':', t1_win)
  print(t2+':', t2_win) 
  return win_by_year
```

##### Batsman Strike rate

```python
def srateDef(name, against, df):
    batsman = df[df['batsman'] == name]
    batsman = batsman[batsman['over']<=6]
    runsdf = batsman[['match_id','batsman_runs','ball']]
    runs = runsdf.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()
    num_balls = runsdf.groupby(['match_id'], sort = False)['ball'].count().reset_index()
    blist = num_balls['ball'].values.tolist()
    rlist = runs['batsman_runs'].values.tolist()
    df2 = pd.DataFrame({'ball':blist,'runs':rlist})
    df2['rate'] = df2['runs']/df2['ball'] * 100
    tot_mat = df2['rate'].describe()[0]
    mean ,median = runs['batsman_runs'].describe()[1], runs['batsman_runs'].describe()[5]
    print(df2['rate'].describe(),'\n')
    if against != '':
        batsman_ag =  batsman[batsman['bowling_team']==against]
        runsdf = batsman_ag[['match_id','batsman_runs','ball']]
        runs = runsdf.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()
        num_balls = runsdf.groupby(['match_id'], sort = False)['ball'].count().reset_index()
        blist = num_balls['ball'].values.tolist()
        rlist = runs['batsman_runs'].values.tolist()
        df2 = pd.DataFrame({'ball':blist,'runs':rlist})
        df2['rate'] = df2['runs']/df2['ball'] * 100
        tot_mat = df2['rate'].describe()[0]
        mean ,median = runs['batsman_runs'].describe()[1], runs['batsman_runs'].describe()[5]
        print(df2['rate'].describe(),'\n')
        print(df2['rate'].value_counts())
```

##### Total score by team

```python
def tot_runs_team(t1,t2, df = deli):
    df1 = df[df['batting_team'] == t1]
    df1 = df1[df1['bowling_team'] == t2]
    tot_run = df1.groupby(['match_id'], sort = False)['total_runs'].sum().reset_index()    
    print(tot_run['total_runs'].describe())
    print(tot_run['total_runs'].sort_values().values)
```

##### Batsman contribution in total score

```python
def bat_cont(bat,t1, against, df):
    new_df = df[df['batting_team']==t1]
    edf = new_df[new_df['bowling_team'] == t1]
    bat_df = new_df[new_df['batsman']==bat]
    for i in bat_df['match_id'].unique():
        edf = pd.DataFrame.append(edf,new_df.iloc[np.where(new_df['match_id'] == i)])
    edf = edf.reset_index().drop(['index'],axis = 1)
    if against != '':
        edf = edf[edf['bowling_team'] == against]
    bat_df = edf[edf['batsman']=='SS Iyer']
    tot_df = edf
    bat_runs = bat_df.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()
    tot_runs = tot_df.groupby(['match_id'], sort = False)['total_runs'].sum().reset_index()
    blist = bat_runs['batsman_runs'].values.tolist()
    rlist = tot_runs['total_runs'].values.tolist()
    df2 = pd.DataFrame({'bat_runs':blist,'tot_runs':rlist})
    df2['percentage'] = df2['bat_runs']/df2['tot_runs'] *100
    print(df2['percentage'].describe(),'\n')
```

##### Wickets by team

```python
def wicket(t1, not_b, df):
    new_df = df[df['bowling_team']==t1]
    not_df = new_df[new_df['bowler']==not_b]
    for i in not_df['match_id'].unique():
        not_df = pd.DataFrame.append(not_df,new_df.iloc[np.where(new_df['match_id'] == i)])
    not_df = not_df[not_df['bowler']!=not_b]
    not_df = not_df[(not_df['over']>=5)&(not_df['over']<=20)]
    tot_wick = not_df.groupby(['match_id'], sort = False)['dismissal_by_bowler'].sum()
    print(tot_wick.describe(),'\n')
    print(tot_wick.value_counts(),'\n\n')
```
