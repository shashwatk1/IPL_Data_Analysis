## Functions

##### Head to Head and last 3 years

```python
def head2head(t1,t2,df = match):
    s1 = (df['team1'] == t1) & (df['team2'] == t2)
    s2 = (df['team1'] == t2) & (df['team2'] == t1)
    head = df.iloc[np.where(s1 | s2)]
    head = head.sort_values(by = 'season').reset_index().drop(['id','index'], axis=1)
    return head

def last3years():
    y1 = deli.iloc[np.where(deli['year'] == 2017)[0]]
    y2 = deli.iloc[np.where(deli['year'] == 2018)[0]]
    y3 = deli.iloc[np.where(deli['year'] == 2019)[0]]
    year3 = pd.DataFrame.append(y1,y2, ignore_index=True)
    year3 = pd.DataFrame.append(year3,y3, ignore_index=True)
    return year3
```

##### Head to Head Score Function

```python
def head(t1, t2):
  head_to_head = head2head(t1,t2)
  t1_win=0
  t2_win=0
  for i in head_to_head['winner']:
    if i == t1:
      t1_win+=1
    if i == t2:
      t2_win+=1
  t1_arr = []
  t2_arr = []
  year_arr = []
  for i in head_to_head[['season','winner']].iterrows():
      index, year, win = i[0], i[1][0], i[1][1]
      if year in year_arr:
          if win == t1:
              t1_arr[year_arr.index(year)]+=1  
          if win == t2:
              t2_arr[year_arr.index(year)]+=1  
      else:
          year_arr.append(year)
          if win == t1:
              t1_arr.append(1)
              t2_arr.append(0)
          if win == t2:
              t1_arr.append(0)
              t2_arr.append(1)
  win_by_year = pd.DataFrame([t1_arr, t2_arr], columns = year_arr, index = [t1, t2])
  print("Head to Head:",t1,'v/s',t2)
  print(t1+':', t1_win)
  print(t2+':', t2_win) 
  return win_by_year
```

##### Toss Function

```python
def toss2fun(df):
    t1_toss = df[df['toss_winner']==t1].reset_index()['toss_decision']
    t2_toss = df[df['toss_winner']==t2].reset_index()['toss_decision']
    toss_df = t1_toss.append(t2_toss).reset_index().drop(['index'], axis=1)
    field = 0
    bat = 0
    for i in toss_df['toss_decision']:
        if i == 'field':
            field+=1
        if i == 'bat':
            bat+=1
    print('Field:',field)
    print('Bat:', bat) 
    
def tossFun(t1, t2):
    print("Toss Prediction:")
    toss2fun(match)
    print("\nToss Prediction Head to Head:")
    toss2fun(head2head(t1,t2))
```

##### Batsman runs calculation function

```python
def runsDef(name, check, against, df):
    c1 = df['batsman'] == name
    c3 = df['bowling_team'] == against
    c0 = df['batsman_runs'] == 0
    batsman = df.iloc[np.where(c1)]
    runsdf = batsman[['match_id','batsman_runs']]
    runs = runsdf.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()
    tot_run, tot_mat = runs['batsman_runs'].sum(), runs['batsman_runs'].describe()[0]
    mean ,median = runs['batsman_runs'].describe()[1], runs['batsman_runs'].describe()[5]
    print('Total =>')
    print('Matches Played: ',int(tot_mat),'\tTotal Runs:',tot_run,'\tMean: ',np.round_(mean, decimals = 3),'\tMedian: ',median)
    if against != '':
        batsman_ag = df.iloc[np.where(c1&c3)]
        runsdf = batsman_ag[['match_id','batsman_runs']].reset_index().drop(['index'], axis=1)
        runs = runsdf.groupby(['match_id'], sort = False)['batsman_runs'].sum().reset_index()
        tot_run, tot_mat = runs['batsman_runs'].sum(), runs['batsman_runs'].describe()[0]
        mean ,median = runs['batsman_runs'].describe()[1], runs['batsman_runs'].describe()[5]
        print("Against", against,'=>')
        print('Matches Played: ',int(tot_mat),'\tTotal Runs:',tot_run,'\tMean: ',np.round_(mean, decimals = 3),'\tMedian: ',median)
    if check != 10:
        print('\nBOUNDARIES')
        if check == 4:
            bound = 'fours:'
            col = 'is_four'
        elif check == 6:
            bound = 'sixes:'
            col = 'is_six'
        else:
            bound = '[default]:'
        batsman_ch = df.iloc[np.where(c1)]
        runsdf = batsman_ch[['match_id',col]].reset_index().drop(['index'], axis=1)
        runs = runsdf.groupby(['match_id'], sort = False)[col].sum().reset_index()
        tot_run, tot_mat = runs[col].sum(), runs[col].describe()[0]
        mean ,median = runs[col].describe()[1], runs[col].describe()[5]
        print('Total =>')
        print('Matches Played:',int(tot_mat),'\tTotal',bound,tot_run,'\tMean:', np.round_(mean, decimals = 3),'\tMedian:',median)
        if against != '':
            batsman_ch_ag = df.iloc[np.where(c1&c3)]
            runsdf = batsman_ch_ag[['match_id',col]].reset_index().drop(['index'], axis=1)
            runs = runsdf.groupby(['match_id'], sort = False)[col].sum().reset_index()
            tot_run, tot_mat = runs[col].sum(), runs[col].describe()[0]
            mean ,median = runs[col].describe()[1], runs[col].describe()[5]
            print("Against", against,'=>')
            print('Matches Played:',tot_mat,'\tTotal',bound,tot_run,'\tMean:', np.round_(mean, decimals = 3),'\tMedian:',median)
```
