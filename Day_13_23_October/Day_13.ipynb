## Functions

##### Head to Head and last 3 years

```python
def head2head(t1,t2,df = match):
    s1 = (df['team1'] == t1) & (df['team2'] == t2)
    s2 = (df['team1'] == t2) & (df['team2'] == t1)
    head = df.iloc[np.where(s1 | s2)]
    head = head.sort_values(by = 'season').reset_index().drop(['id','index'], axis=1)
    return head

y1 = deli.iloc[np.where(deli['year'] == 2017)[0]]
y2 = deli.iloc[np.where(deli['year'] == 2018)[0]]
y3 = deli.iloc[np.where(deli['year'] == 2019)[0]]
year3 = pd.DataFrame.append(y1,y2, ignore_index=True)
year3 = pd.DataFrame.append(year3,y3, ignore_index=True)
last3years = year3
```

##### Head to Head Score Function

```python
def head(t1, t2):
  head_to_head = head2head(t1,t2)
  t1_win=0
  t2_win=0
  for i in head_to_head['winner']:
    if i == t1:
      t1_win+=1
    if i == t2:
      t2_win+=1
  t1_arr = []
  t2_arr = []
  year_arr = []
  for i in head_to_head[['season','winner']].iterrows():
      index, year, win = i[0], i[1][0], i[1][1]
      if year in year_arr:
          if win == t1:
              t1_arr[year_arr.index(year)]+=1  
          if win == t2:
              t2_arr[year_arr.index(year)]+=1  
      else:
          year_arr.append(year)
          if win == t1:
              t1_arr.append(1)
              t2_arr.append(0)
          if win == t2:
              t1_arr.append(0)
              t2_arr.append(1)
  win_by_year = pd.DataFrame([t1_arr, t2_arr], columns = year_arr, index = [t1, t2])
  print("Head to Head:",t1,'v/s',t2)
  print(t1+':', t1_win)
  print(t2+':', t2_win) 
  return win_by_year
```

##### Batsman runs by boundaries in the first n balls

```python
def bat_bound(ba,n,against,df):
    bat_bound_runs = []
    mat = []
    bat = df[df['batsman']==ba]
    if against != '':
        bat= bat[bat['bowling_team']==against]
    for i in bat['match_id'].unique():
        bat_mat = bat[bat['match_id']==i]   
        if n != 0:
            bat_mat = bat_mat.iloc[:n]
        bat_bound_runs.append(bat_mat['bound_runs'].sum())
        mat.append(i)
    df2 = pd.DataFrame(bat_bound_runs, columns = ['runs'])
    print(df2['runs'].value_counts(),'\n')
    print(df2['runs'].describe())
```

##### Bowler dots to total ball ratio

```python
def dt_tot_ratio(bowl, against, df, col = 0):
    col_name = 'is_dot_1'
    if col != 0:
        col_name = 'is_dot_2'
    df = df[df['bowler']==bowl]
    if against!= '': 
        df = df[df['batting_team']==against]
    dots = []
    total = []
    for i in df['match_id'].unique():
        ball = df[df['match_id']==i]
        total.append(ball['ball'].count())
        dots.append(ball[col_name].sum())
    df2 = pd.DataFrame({'tot_ball':total,'dot_ball':dots})
    df2['ratio'] = df2['dot_ball']/df2['tot_ball']
    count, mean ,median = df2['ratio'].describe()[0], df2['ratio'].describe()[1], df2['ratio'].describe()[5]
    print('Matches Played: ',count,'\tMean: ',np.round_(mean, decimals = 3),'\tMedian: ',np.round_(median, decimals = 3))
```

##### Overs with 4 and 6 both

```python
def overbound(t1, t2, df):
    newdf = df[((df['batting_team']==t1)&(df['bowling_team']==t2))|((df['batting_team']==t2)&(df['bowling_team']==t1))]
    newdf = newdf.groupby(['match_id','over'], sort = False)[['is_four','is_six']].sum().reset_index()
    newdf = newdf[(newdf['is_four']>=1)&(newdf['is_six']>=1)]
    newdf = newdf.groupby(['match_id'], sort = False)['over'].count().reset_index()
    print(newdf['over'].value_counts(),'\n')
    print(newdf['over'].describe())
```

##### Bowler lethality score

```python
    def bls_score(name, against, df):
        wick, dot, bound_count, bound_run, tot = [], [], [], [], []
        bdf = df[df['bowler']==name]
        bdf = bdf[bdf['batting_team']==against]
        for i in bdf['match_id'].unique():
            bdf2 = bdf[bdf['match_id']==i]
            wick.append(bdf2.groupby(['match_id'], sort = False)['dismissal_by_bowler'].sum().reset_index()['dismissal_by_bowler'][0])
            dot.append(bdf2.groupby(['match_id'], sort = False)['is_dot_2'].sum().reset_index()['is_dot_2'][0])
            bound_count.append(bdf2.groupby(['match_id'], sort = False)['is_bound'].sum().reset_index()['is_bound'][0])
            bound_run.append(bdf2.groupby(['match_id'], sort = False)['bound_runs'].sum().reset_index()['bound_runs'][0])
            tot.append(bdf2.groupby(['match_id'], sort = False)['ball'].count().reset_index()['ball'][0])
        ndf = pd.DataFrame({'tot_balls':tot,'dot_balls':dot,'bound_count':bound_count,'bound_run':bound_run,'wickets':wick})  
        ndf['BLS'] = np.round_(((ndf['wickets'] + ndf['dot_balls']- (ndf['bound_run']/ndf['bound_count']))/(ndf['tot_balls'])), decimals = 2)
        count, mean ,median = ndf['BLS'].describe()[0], ndf['BLS'].describe()[1], ndf['BLS'].describe()[5]
        print('Matches Played: ',count,'\tMean: ',np.round_(mean, decimals = 3),'\tMedian: ',np.round_(median, decimals = 3))
```
